```python
import cv2
from brisque import BRISQUE
from PIL import Image

def evaluate_brisque_score(file_path):
    image = Image.open(file_path).convert("RGB")
    quality_assessor = BRISQUE()
    quality_score = quality_assessor.get_score(image)
    return quality_score

# Example usage
sample_image_path = "example.jpg"
score = evaluate_brisque_score(sample_image_path)
print("BRISQUE Score:", score)
```
