### **Importing Libraries**

```python
import numpy as np
import pandas as pd
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from scipy import stats
from collections import Counter
from sklearn.decomposition import TruncatedSVD
import matplotlib.pyplot as plt


#Loading Training File
from glob import glob
import os
import pandas as pd


allColumns = [1,'forward','head_pos_x','head_pos_y','eye','gaze_pos_x','gaze_pos_y']
reqColumns = ['head_pos_x','head_pos_y','gaze_pos_x','gaze_pos_y','participant']

```

### **Choosing 30 vedios of each user** 

```python
df_array = []

for i in range(1,10):
  files = glob(f'data/Gaze_txt_files_2/p00{i}/*')
  files.sort()
  files = files[:30]

  tempDF = pd.concat( ( pd.read_csv(file,header=None, float_precision='high',names = allColumns) for file in files) , ignore_index = True)
  tempDF = tempDF.assign(participant=pd.Series(i, index=tempDF.index))
  tempDF = tempDF.loc[:, reqColumns]
  print(f"{i} particcipant have record{tempDF.shape}")
  df_array.append(tempDF)


for i in range(10,46):
  files = glob(f'data/Gaze_txt_files_2/p0{i}/*')
  files.sort()
  files = files[:30]

  tempDF = pd.concat( ( pd.read_csv(file,header=None, float_precision='high',names = allColumns) for file in files) , ignore_index = True)
  tempDF = tempDF.assign(participant=pd.Series(i, index=tempDF.index))
  tempDF = tempDF.loc[:, reqColumns]
  print(f"{i} particcipant have record{tempDF.shape}")
  df_array.append(tempDF)

dataSet = pd.concat( ( dfs for dfs in df_array) , ignore_index = True)
```

## Storing data from array to a csv file 

```python
dataSet.to_csv("data/final_data.csv", index = False)
```
